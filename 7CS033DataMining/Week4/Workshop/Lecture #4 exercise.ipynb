```python
import pandas as pd
import numpy as np
```

```python
# Replace the data path with the data path on your local disk
data=pd.read_csv('C:/Users/u27408/Downloads/cancer.csv')
```

```python
data
```

```python
X=data.drop('Diagnosis',axis=1)
y=data['Diagnosis']
```

```python
# preprocess data
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(X) # normalizing the features
#PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
X_df = pd.DataFrame(data = X_pca, columns = ['principal component 1', 'principal component 2'])
```

```python
X_df
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component - 1',fontsize=20)
plt.ylabel('Principal Component - 2',fontsize=20)
plt.title("Principal Component Analysis of the Cancer dataset",fontsize=20)
targets_Diagnosis = ['M', 'B']
colors = ['r', 'y']
for target, color in zip(targets_Diagnosis,colors):
    indicesToKeep = data['Diagnosis'] == target
    plt.scatter(X_df.loc[indicesToKeep, 'principal component 1']
               , X_df.loc[indicesToKeep, 'principal component 2'], c = color, s = 50)

plt.legend(targets_Diagnosis,prop={'size': 15})
```
