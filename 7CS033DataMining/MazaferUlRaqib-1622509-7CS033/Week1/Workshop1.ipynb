* Q10: Write a function that receives the dataset and replace Famle with F and Male with M (please try to write it yourself before checking the answer in the next cell)

```python
# Answer
def encode_sex(data):
    rows=data.shape[0]
    a=data.loc[:,'sex']
    for i in range(rows): 
        if a[i]=="Male":
            data.loc[i,"sex"]="M"
        elif a[i]=="Female":
            data.loc[i,"sex"]="F"
    return data
```

```python
# Copying the data
data_copy=data.copy()
```

## TASK 1.1: Import the dataset from the URL we used in this workshop. Then generate a new dataset by randomly extracting 10000 samples. Reindex the generated dataset and remove NULL values. Name the new dataset 'task_dataset'

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/maz786/WlvMScAI/main/7CS033DataMining/Week1/Workshop/adult.csv'

data = pd.read_csv(url)

task_dataset = data.sample(n=10000, random_state=42)

task_dataset.reset_index(drop=True, inplace=True)

task_dataset.dropna(inplace=True)

print(task_dataset.head())
```

## TASK 1.2: Write a code to find how much contribution each sex and occupation category made to the capital-gain on average. Apply the code to the task_dataset and report the result (Hint: you need to use the groupby method)

```python
average_capital_gain = task_dataset.groupby(['sex', 'occupation'])['capital-gain'].mean()

print(average_capital_gain)
```

## TASK 1.3: Write a code to find the country with the highest number of people with a Bachelors degree. Apply the code to the task_dataset and report the result (Hint: you need to use the groupby method)

```python
bachelors_data = task_dataset[task_dataset['education'] == 'Bachelors']

country_counts = bachelors_data.groupby('native-country').size()

highest_bachelors = country_counts.idxmax()

print(f"The country with the highest number of people with a Bachelor's degree is: {highest_bachelors}")
```

## TASK 1.4: Write a code to receive two lists including five names and their respective ages and print 'Hello Name Age' 
## For example, if it received a list of two names ['Amin', 'Michael'] and respective ages [27,38], it would print 'Hello Amin 27', 'Hello Michael 38'. Each hello statement should be printed in a new line 

```python
def print_greetings(names, ages):
    for name, age in zip(names, ages):
        print(f"Hello {name} {age}")

names = ['Amin', 'Michael']
ages = [27, 38]
print_greetings(names, ages)
```

## TASK 1.6: Write a function to split the task_dataset in half column-wise and swap the first half and the second half

```python
def swap_dataset_columns(data):
    midpoint = data.shape[1] // 2

    first_half = data.iloc[:, :midpoint]
    second_half = data.iloc[:, midpoint:]

    swapped_data = pd.concat([second_half, first_half], axis=1)

    return swapped_data

swapped_task_dataset = swap_dataset_columns(task_dataset)

print(swapped_task_dataset.head())
```

## Task 1.7: Write a function that receives two numerical columns' names and compare their values for all rows. If the value of the first column is greater than the second column, the function should produce True, otherwise, it should produce False. The function should append an additional column to the dataset to store the results of the comparison for all rows. Apply the function to the "age" and "hours-per-week" columns in the task_dataset and print the result.

```python
def compare_columns_and_append(data, col1, col2, new_col_name):
    comparison_result = data[col1] > data[col2]

    data[new_col_name] = comparison_result

compare_columns_and_append(task_dataset, 'age', 'hours-per-week', 'age_greater_than_hours_per_week')

print(task_dataset[['age', 'hours-per-week', 'age_greater_than_hours_per_week']].head())
```

## Task 1.8: Write a function that returns the names of countries with maximum and minimum average ages in the task_dataset (Hint: you can use the numpy module)

```python
import numpy as np

def find_countries_with_extreme_ages(data):
    average_ages = data.groupby('native-country')['age'].mean()

    max_age_country = average_ages.idxmax()
    max_age_value = np.max(average_ages)

    min_age_country = average_ages.idxmin()
    min_age_value = np.min(average_ages)

    return max_age_country, max_age_value, min_age_country, min_age_value

max_age_country, max_age_value, min_age_country, min_age_value = find_countries_with_extreme_ages(task_dataset)

print(f"Country with maximum average age: {max_age_country} (Average Age: {max_age_value})")
print(f"Country with minimum average age: {min_age_country} (Average Age: {min_age_value})")
```
