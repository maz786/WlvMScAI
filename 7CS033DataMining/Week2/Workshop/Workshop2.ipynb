```python
import pandas as pd

url = 'https://raw.githubusercontent.com/maz786/WlvMScAI/main/7CS033DataMining/Week2/Workshop/adult.csv'

df = pd.read_csv(url)

df_cleaned = df.dropna()

null_values_after_cleaning = df_cleaned.isna().sum()

print(null_values_after_cleaning)
```

**<u>Task 2.2 c)</u>**

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/maz786/WlvMScAI/main/7CS033DataMining/Week2/Workshop/adult.csv'

df = pd.read_csv(url)

categorical_columns = df.select_dtypes(include=['object']).columns

categorical_columns_with_null = [col for col in categorical_columns if df[col].isna().any()]

for column in categorical_columns_with_null:
    mode_value = df[column].mode()[0]  # Find the mode
    df[column].fillna(mode_value, inplace=True)  # Fill null values with the mode

null_values_after_filling = df[categorical_columns_with_null].isna().sum()

print(null_values_after_filling)
```

In accordance with best practices in data preprocessing for improved data integrity and analysis quality, filling Null values with the mode for categorical columns ('workclass, occupation, and native-country') effectively handles missing data (Hair et al., 2019).

## TASK 2.3. Select one of the columns which contains wrong data. Write a code to fill the wrong cells with an appropriate value. You have the freedom to determine what value you want to use to fill the wrong cells (2%). 

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/maz786/WlvMScAI/main/7CS033DataMining/Week2/Workshop/adult.csv'

df = pd.read_csv(url)

initial_count = (df['native-country'] == '?').sum()

df['native-country'] = df['native-country'].replace('?', 'Diverse Origins')

final_count = (df['native-country'] == '?').sum()

print(f"Initial count of '?' in 'native-country': {initial_count}")
print(f"Final count of '?' in 'native-country' after replacement: {final_count}")
```

This method supports the idea that data should be as rich and diverse as its sources are, which deepens demographic analysis (Zhang, 2016).

**<u>Task 2.4 a)</u>**

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/maz786/WlvMScAI/main/7CS033DataMining/Week2/Workshop/adult.csv'

df = pd.read_csv(url)

df['income'] = df['income'].str.replace('\.', '', regex=True)

df['income_encoded'] = df['income'].map({'<=50K': 0, '>50K': 1})

print(df[['income', 'income_encoded']].head())
```

The 'income' column encoding for binary classification highlights how important data pretreatment is to machine learning. We enable interpretability of the model and computational efficiency by mapping revenue levels to binary values. This procedure improves model performance and forecast accuracy by demonstrating crucial data preparation techniques (Hastie et al., 2009).

**<u>Task 2.4 b)</u>**

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/maz786/WlvMScAI/main/7CS033DataMining/Week2/Workshop/adult.csv'

df = pd.read_csv(url)

education_order = {
    'Preschool': 0, '1st-4th': 1, '5th-6th': 2, '7th-8th': 3, '9th': 4,
    '10th': 5, '11th': 6, '12th': 7, 'HS-grad': 8, 'Some-college': 9,
    'Assoc-voc': 10, 'Assoc-acdm': 11, 'Bachelors': 12, 'Masters': 13,
    'Prof-school': 14, 'Doctorate': 15
}

df['education_encoded'] = df['education'].map(education_order)

print(df[['education', 'education_encoded']].sample(10))
```

The qualitative order of educational achievement is effectively converted into a numeric scale by ordinal encoding of the 'education' column. This approach maintains the natural hierarchy, which is essential for models to understand the evolution and use it as a predictor (Agresti, 2010).

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

url = 'https://raw.githubusercontent.com/maz786/WlvMScAI/main/7CS033DataMining/Week2/Workshop/adult.csv'
df = pd.read_csv(url)

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

scaler = MinMaxScaler()

df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

print(df[numerical_cols].head())
```

Using MinMaxScaler, which scales each feature to a specified range—in this case, 0 to 1—normalization of the dataset's numerical columns was accomplished. In order to ensure equal contribution across features, this preprocessing step is crucial for models that are sensitive to input scale and variance. According to Jain et al. (2005), normalisation improves both the predictive performance and the efficiency of training models by treating all numerical variables consistently.
