### Step-by-Step Code:

```python
import pandas as pd

file_path = '/Users/maz/7CS033DataMining/Week2/Workshop/adult.csv'

adult_data = pd.read_csv(file_path)

categorical_columns_with_nulls = ['workclass', 'occupation', 'native-country']

print("Initial Null Values Count:")
print(adult_data[categorical_columns_with_nulls].isna().sum())

for column in categorical_columns_with_nulls:
    mode_value = adult_data[column].mode()[0]  
    print(f"Mode for {column}: {mode_value}")  
    adult_data[column].fillna(mode_value, inplace=True)  

print("\nNull Values Count After Filling:")
print(adult_data[categorical_columns_with_nulls].isna().sum())
```

## TASK 2.3. Select one of the columns which contains wrong data. Write a code to fill the wrong cells with an appropriate value. You have the freedom to determine what value you want to use to fill the wrong cells (2%). 

In TASK 2.3, we will select a column containing erroneous data and determine the appropriate approach to replace values that violate the law of consistency or contain errors. An excellent choice for the workclass column, based on our recent conversation, is the placeholder value "?", which we identified as signifying missing or erroneous data. 

Since using the mode is a standard procedure for addressing missing categorical data, let's assume that we would like to replace these placeholder values with the mode value of the data within the workclass column. This method makes the assumption that the mode, particularly in the lack of more evidence to infer a more precise fill value, is the most likely category for missing data.

Step-by-Step Code: 

- **Determine the Mode: In the "workclass" column, find the most frequently occurring value where the placeholder question mark is equivalent to "?".**
- **Replace Placeholders:** Substitute the discovered potential mode for the "?" values.
- **Verify the Replacement:** After the replacement, count the number of times "?" appears to ensure that it was successful.

Here's how you can do that:

```python
import pandas as pd

file_path = '/Users/maz/7CS033DataMining/Week2/Workshop/adult.csv'

adult_data = pd.read_csv(file_path)

mode_value = adult_data.loc[adult_data['workclass'] != '?', 'workclass'].mode()[0]

adult_data['workclass'] = adult_data['workclass'].replace('?', mode_value)

remaining_placeholders = (adult_data['workclass'] == '?').sum()

print(f"Mode value used for replacement in 'workclass': {mode_value}")
print(f"Count of '?' placeholders after replacement in 'workclass': {remaining_placeholders}")
```

### <span style="font-size: 16.38px;">Finding Mode: To make sure the value in the workclass column truly relates to a category, we first identify the value that appears the most frequently in the column by eliminating any "?" values.&nbsp;</span> 

### <span style="font-size: 16.38px;">Replacing Placeholders: In the next step, we enter this mode value in lieu of every "?" that appears in the workclass column. The replace() method is used to accomplish this.</span>

### <span style="font-size: 16.38px;">Verification: In the end, we count the number of "?" values that remain in the workclass column to determine the success of the replacement. If all of the replacements were successful, this number should be 0.&nbsp;</span>

Label encoding is a popular technique for encoding binary categorical variables, in which a distinct number is assigned to each category. **<u>2.4 c)</u>**

One-hot encoding is a suitable and often utilised technique for encoding a nominal column without suggesting any hierarchy among its categories. Models can handle each category as an independent feature without assuming any order by using this strategy, which converts each categorical value into a new binary column. 

Let's choose to use one-hot encoding and encode the nominal column workclass. To do this, we'll use the simple and effective get\_dummies() method in pandas, which creates one-hot encoded variables.  

### Step-by-Step Code:

```python
import pandas as pd

file_path = '/Users/maz/7CS033DataMining/Week2/Workshop/adult.csv'
adult_data = pd.read_csv(file_path)

workclass_encoded = pd.get_dummies(adult_data['workclass'], prefix='workclass')

workclass_encoded.head()
```

With the help of this code snippet, a new DataFrame called workclass\_encoded will be created, with each workclass category represented as a distinct column that is prefixed with workclass\_. For every row, the column that corresponds to its category will have a 1; all other columns will have 0s. 

The rationale behind pd.get\_dummies() The workclass column in the adult\_data table is the target of this function. In order to identify the source column (workclass) after encoding, we apply a prefix to the new columns using the prefix option. 

The workclass\_encoded as a result For every category in the original workclass column, DataFrame includes binary columns. This successfully transforms the nominal data into a format that is appropriate for machine learning algorithms without suggesting any kind of ordinal relationship.  

When handling nominal data for machine learning preprocessing, this approach is very helpful because it guarantees that the model can distinguish between different categories without making any unintentional ordinal assumptions.

**<u>2.4 e)</u>**

There are several methods you can use to normalise the dataset's numerical columns. Using Min-Max normalisation, which scales the data in each column to a given range (often \[0, 1\]), is a popular method. With the help of the MinMaxScaler from sklearn.preprocessing, this is readily accomplished. 

In machine learning, normalisation is especially helpful in preventing bias resulting from feature scale by ensuring that numerical columns contribute equally to the model's training process. 

The following steps will help you normalise the dataset's numerical columns: 

Step 1, Determine the Numerical Columns. Determine which of your dataset's columns are numerical first. Number columns such as age, fnlwgt, education-num, capital-gain, capital-loss, and hours-per-week are common examples, assuming the Adult dataset. 

Step 2: Use Normalisation (Min-Max)

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

file_path = '/Users/maz/7CS033DataMining/Week2/Workshop/adult.csv'  

adult_data = pd.read_csv(file_path)

numerical_columns = ['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']

scaler = MinMaxScaler()

adult_data[numerical_columns] = scaler.fit_transform(adult_data[numerical_columns])

adult_data.head()
```

### <span style="font-size: 16.38px;">Reasoning: Recognising Numerical Columns Normalisation should only be applied to the numerical columns because scaling categorical data in this manner would be illogical and might generate mistakes in the data.&nbsp;</span> 

### <span style="font-size: 16.38px;">MinMaxScaler: This sklearn utility.Preprocessing is the act of transforming and scaling the data so that each attribute in the input data is scaled to fall inside one of two ranges: [0, 1] or an otherwise specified range.&nbsp;</span> 

### <span style="font-size: 16.38px;">Fitting and Transforming: To scale the data to the designated range, the fit_transform() method calculates the minimum and maximum values of the data to be changed. You may guarantee that the values in your dataset are confined inside a particular range by normalising the numerical columns. This will enable more stable and rapid convergence while training machine learning models.</span>
