We will apply the apriori algorithm to France's online retail data

Please see the following pages for further instructions on how to use the apriori module

https://rasbt.github.io/mlxtend/user_guide/frequent_patterns/apriori/

https://rasbt.github.io/mlxtend/user_guide/frequent_patterns/association_rules/


```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
basket_France = pd.read_csv('/content/drive/MyDrive/Wlv/7CS033/France_online_retail.csv')
```

```python
basket_France
```

```python
basket_France=basket_France.drop('InvoiceNo',axis=1)
```

```python
# Finding the frequent itemsets
frq_items = apriori(basket_France, min_support = 0.1, use_colnames = True)
```

```python
frq_items.sort_values(by=['support'],ascending=False)
```

```python
# Collecting the inferred rules in a dataframe
rules = association_rules(frq_items, metric ="support", min_threshold = 0.1)
```

```python
rules.sort_values(by=['support'],ascending=False)
```
